# ROS2 Robot Navigation

## Navigate the robot and execute specific tasks

To make the assignment more accessible, a simple robot simulator has been implemented
to avoid running the resource intensige Gazebo simulator.
The goal is to replicate the output of the screenshot below:
(this is a screenshot of our custom simulator)
<img src='demo_outcome.png'/>

### Notes
* The purple rectangle is the robot's navigation path* The red 'n' yellow face is the robot* The path's rectangle size does not matter, but it does not go through the center of the map
* The robot ends up in the center
* The robot's path is not connected with the robot itself
* The robot is happy!


Let's execute all the required commands

Open a terminal and run the custom simulation node:
```bash
ros2 run ai_msc_robotics_assignments_ros robot_sim.py
```

Use the web video server node to expose the ROS image of the "simulator" to a stream, viewable in browser: Open a terminal and run the python node:
```bash
ros2 run web_video_server web_video_server
```


Visit the following link to see the input image data in your browser:
<a href='http://localhost:8080/stream?topic=/robot_sim_visualization' target='_blank'>http://loca:8080/stream?topic=/robot_sim_visualization</a>

You are now ready to start coding. (No code is provided for this assignment)
